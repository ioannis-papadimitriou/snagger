```python
# Now also plot the mean for each different feature
print("Now also plot the mean for each different feature")
t = dataset.target[:]
n_elem = 10.0
for i in range(dataset.data.shape[1]):
    print("Feature %d (%s)" % (i, dataset.feature_names[i]))
    d = dataset.data[:, i]
    step = (np.max(d) - np.min(d)) / n_elem
    r = np.arange(np.min(d) - step / 2, np.max(d) + 3 * step / 2, step)
    v = np.zeros(r.shape[0] - 1)
    for j in range(r.shape[0] - 1):
        idx = np.logical_and(d >= r[j], d < r[j + 1])
        if (t[idx]).size == 0:
            v[j] = np.nan
        else:
            v[j] = np.mean(t[idx])
    plt.plot(d, t, 'x')
    ax = plt.gca()
    ax.locator_params(axis='both', nbins=10)
    plt.plot(r[:-1] + step / 2, v, 'or', markersize=12)
    plt.xlabel(dataset.feature_names[i])
    plt.ylabel("Target")
    plt.grid()
    plt.show()
```

```python
# Let's start building and testing some models

# First, shuffle the data (depending on the use-case, this may not be needed)
idx = np.arange(dataset.data.shape[0])
np.random.shuffle(idx)
dataset.data[:, :] = dataset.data[idx, :]
dataset.target[:] = dataset.target[idx]
```

```python
# Split into training and test set
train_id = int(dataset.data.shape[0] * 0.7)
X_train = dataset.data[0:train_id, :]
X_test = dataset.data[train_id:, :]
y_train = dataset.target[0:train_id]
y_test = dataset.target[train_id:]
```

```python
# Ordinary Least Squares Linear Regression
lr = lm.LinearRegression().fit(X_train, y_train)
# Predict
yhat = lr.predict(X_test)
mse_ols = np.mean((y_test - yhat) ** 2)
mae_ols = np.mean(np.fabs(y_test - yhat))
print("MSE: %f" % (mse_ols))
print("MAE: %f" % (mae_ols))
average_y = np.mean(y_test)
mae_perc = 100.0 * mae_ols/average_y
print("Average price in test set: %f\nError percentage: %f %%\n" % (average_y, mae_perc))
# Vector w=[w_1, w_2, ..., w_p]
print(lr.coef_)
# Intercept term, w_0
print(lr.intercept_)
```

```python
# Ridge regression
ridge = lm.Ridge(alpha=200.0)
ridge.fit(X_train, y_train)
yhat = ridge.predict(X_test)
mse_ridge = np.mean((y_test - yhat) ** 2)
mae_ridge = np.mean(np.fabs(y_test - yhat))
print("MSE: %f" % (mse_ridge))
print("MAE: %f" % (mae_ridge))
mae_perc = 100.0 * mae_ridge/average_y
print("Average price in test set: %f\nError percentage: %f %%\n" % (average_y, mae_perc))
# Vector w=[w_1, w_2, ..., w_p]. Ridge regression results in smaller weights.
print(ridge.coef_)
# Intercept term, w_0
print(ridge.intercept_)
```

```python
# LASSO regression
lasso = lm.Lasso(alpha=10)
lasso.fit(X_train, y_train)
yhat = lasso.predict(X_test)
mse_lasso = np.mean((y_test - yhat) ** 2)
mae_lasso = np.mean(np.fabs(y_test - yhat))
print("MSE: %f" % (mse_lasso))
print("MAE: %f" % (mae_lasso))
mae_perc = 100.0 * mae_lasso/average_y
print("Average price in test set: %f\nError percentage: %f %%\n" % (average_y, mae_perc))
# Vector w=[w_1, w_2, ..., w_p]. LASSO results in sparse weight vector.
print(lasso.coef_)
# Intercept term, w_0
print(lasso.intercept_)
```

```python
# We can do the same with multiple iterations
n_iter = 10
mse = np.zeros(n_iter)
mae = np.zeros(n_iter)
for i in range(n_iter):
    #    np.random.seed(int(time.time()))
    X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(
        dataset.data, dataset.target, test_size=0.3)
    # Train
    lr = lm.LinearRegression().fit(X_train, y_train)
    # Predict
    yhat = lr.predict(X_test)
    mse[i] = np.mean((y_test - yhat) ** 2)
    mae[i] = np.mean(np.fabs(y_test - yhat))
#    mse2[i] = sk.metrics.mean_squared_error(y_test, yhat)

mse_ols = np.mean(mse)
mse_ols_std = np.std(mse)
mae_ols = np.mean(mae)
mae_ols_std = np.std(mae)
print("Average over " + str(n_iter) + " runs:")
print("MSE: %f (%f) " % (mse_ols, mse_ols_std))
print("MAE: %f (%f) " % (mae_ols, mae_ols_std))
```

```python
### SVR with GridSearchCV

from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
from sklearn import metrics
from sklearn.model_selection import GridSearchCV, cross_validate

param_grid = [
  {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
  {'C': [1, 10, 100, 1000], 'gamma': [0.0001, 0.001, 0.005, 0.1, 1, 3, 5],'epsilon': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10],
   'kernel': ['rbf']},
  {'C': [0.1, 1, 100, 1000], 'epsilon': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10], 'degree': [2, 3, 4],
   'coef0': [0.1, 0.01, 0.001, 0.0001], 'kernel': ['poly']}
]

X_s = dataset.data
y_s = dataset.target
# Feature Scaling
scaler = MinMaxScaler(feature_range=(0, 1))
X_s = scaler.fit_transform(X_s)
    
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_s, y_s, test_size = test_size, random_state = 0)

# Fitting SVR to the dataset
svr = SVR()
clf = GridSearchCV(svr, param_grid,cv=5, scoring='neg_mean_squared_error', verbose = 3, n_jobs = 4)
clf.fit(X_train_s, y_train_s)

# Predicting the Test set results
y_pred_s = clf.predict(X_test_s)

plt.figure(figsize=(13,6))
plt.scatter(y_test_s, y_pred_s, c = 'green') 
plt.xlabel("Price: in $1000's") 
plt.ylabel("Predicted value") 
plt.title("True value vs predicted value : Support Vector Regression") 
plt.show()

# The mean squared error
print('\nThe RMSE is: %f' %rmse(y_test_s, y_pred_s))
mae_ols_s = np.mean(np.fabs(y_test_s - y_pred_s))
average_y_s = np.mean(y_test_s)
mae_perc_s = 100.0 * mae_ols_s/average_y_s
print("Average price in test set: %f\nError percentage: %f %%\n" % (average_y_s, mae_perc_s))

print(clf.best_params_)
```
